```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8")
```

```python
# Вивести перші рядки таблиці за допомогою методу head
data_frame = data[12]
data_frame.head()
```

```python
# Визначте кількість рядків та стовпців у датафреймі (атрибут shape)
data_frame.shape
```

```python
# Замініть у таблиці значення "—" на значення NaN
data_frame.replace('—', pd.NA, inplace=True)
data_frame
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
data_frame.dtypes
```

```python
# Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"
df_only_numeric = data_frame.apply(pd.to_numeric, errors='ignore')
df_only_numeric.dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
data_frame.isnull().sum()
```

```python
# Видаліть з таблиці дані по всій країні, останній рядок таблиці
df_only_numeric = df_only_numeric.drop(27)
df_only_numeric
```

```python
# Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)
df_fill_na = df_only_numeric.fillna(df_only_numeric.mean(axis=0, numeric_only=True))
df_fill_na
```

```python
# Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні
average_year_data = df_fill_na.describe()['2019']['mean']
average_year_data

```

```python
#У якому регіоні була найвища народжуваність у 2014 році?
max_data = df_fill_na['2014'].max()
average_reg_data = df_fill_na[df_fill_na['2014'] == max_data]['Регіон'].iloc[0]
average_reg_data
```

```python
#Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

plt.bar(df_only_numeric['Регіон'], df_fill_na['2019'], width = 0.75)
plt.xlabel('Регіон', fontsize="small")
plt.ylabel('Рівень народжуваності у 2019 році', fontsize="small")
plt.title('Народжуваність по регіонах у 2019 році')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df_fill_na
fig, axs = plt.subplots(2, 2, constrained_layout=True, figsize=(8, 6))
axs[0][0].plot(df['1950'], label='1950', color='green', linestyle='--')
axs[0][0].set_title('1950')
axs[0][1].plot(df['1970'], label='1970', color='orange', linestyle='-.')
axs[0][1].set_title('1970')
axs[1][0].plot(df['2000'], label='2000', color='blue', linestyle=':')
axs[1][0].set_title('2000')
axs[1][1].plot(df['2019'], label='2019', color='red', linestyle='-')
axs[1][1].set_title('2019')

plt.suptitle("Коефіцієнт народжуваності в Україні", color='red', fontsize=16)

plt.legend(labels=df.columns)


plt.show()
```
