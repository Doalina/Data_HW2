```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#Прочитайте файл 2017_jun_final.csv за допомогою методу read_csv
data = pd.read_csv('2017_jun_final.csv')
```

```python
# Прочитайте отриману таблицю, використовуючи метод head
data.head()
```

```python
# Визначте розмір таблиці за допомогою методу shape
data.shape
```

```python
# Визначте типи всіх стовпців за допомогою dataframe.dtypes
pd.DataFrame(data).dtypes
```

```python
# Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)
data.isnull().sum()
```

```python
# Видаліть усі стовпці з пропусками, крім стовпця "Мова програмування"
columns_data = ['Язык.программирования'] + list(data.dropna(axis=1, how='any').columns)
clean_data = data[columns_data]
clean_data
```

```python
# Знову порахуйте, яка частка пропусків міститься в кожній колонці і переконайтеся, що залишився тільки стовпець "Мова.програмування"
clean_data.isnull().sum()
```

```python
# Видаліть усі рядки у вихідній таблиці за допомогою методу dropna
clean_data = clean_data.dropna()
clean_data
```

```python
# Визначте новий розмір таблиці за допомогою методу shape
clean_data.shape
```

```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data =  clean_data[clean_data["Язык.программирования"] == "Python"]
python_data
```

```python
# Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"
position = clean_data.groupby("Должность", as_index=False)
position
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
salary  = position.agg({"Зарплата.в.месяц": ("min","max")})
salary
```

```python
# Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. 
# Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(row):
    min_salary = row['Зарплата.в.месяц']['min']
    max_salary = row['Зарплата.в.месяц']['max']
    return np.average([min_salary, max_salary])


salary['avg'] = salary.apply(fill_avg_salary, axis=1)
salary
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
salary["avg"].describe(include='all')
```

```python
#Збережіть отриману таблицю в CSV файл
save_salary = salary.to_csv("salary.csv", index=True)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.violinplot(x='Зарплата.в.месяц', data=data)

plt.xlabel('Зарплата в месяц', fontsize=12)
plt.title('Распределение зарплат в месяц', fontsize=14)

plt.xlim(right=8000)

plt.show()
```

```python
sns.histplot(data=data, x='Изменение.зарплаты.за.12.месяцев', kde=True)
plt.xlabel('Изменение зарплаты')
plt.ylabel('Частота')
plt.title('Изменение зарплаты за 12 месяцев')

plt.xlim(left=-500, right=2000)
plt.show()
```

```python
sns.boxplot(x='Город', y='Зарплата.в.месяц', data=data)
plt.xlabel('Город')
plt.ylabel('Зарплата в месяц')
plt.title('Распределение зарплат по городам')
plt.xticks(rotation=90)
plt.show()
```
